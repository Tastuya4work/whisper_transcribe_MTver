```python
"""
←左の再生（▷）ボタンを押して実行
"""
!pip install git+https://github.com/openai/whisper.git >/dev/null
# !pip install git+https://github.com/openai/whisper.git@28769fcfe50755a817ab922a7bc83483159600a9 >/dev/null
import os
import whisper
model = whisper.load_model("large")
```

# ④プログラムの実行_2（実行が必要）
実行前に必ず音声データのアップロードを実施してください。

出力として、文字起こしデータが出てきます。

```python
"""
←左の再生（▷）ボタンを押して実行
"""

f_list = os.listdir(path='.')

for files in f_list:
  if "m4a" in files or "mp3" in files:
    print("%sを文字起こしします" %files)
    result = model.transcribe( audio = "/content/%s" % files, verbose=True, language="ja")
    print(result["text"])
```

```python
"""
AIによる要約を簡易に実行できるように変形する。
現在開発中
"""
import textwrap
chunks = textwrap.wrap(result, 2500)

for words in chunks:
  print(words)
  print("\r\n\r\n\r\n")
```
